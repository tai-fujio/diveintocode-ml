# 【課題1】練習のために特徴量とカテゴリを選択
* irisデータセットをWeek3同様に取得した後、データフレームから下記を抜き出す
    - virgicolorとvirginica(1・2)

```python
from sklearn.datasets import load_iris
import pandas as pd

# irisデータセットをロード
iris = load_iris()

# 説明変数をdf型に変換し、カラムをセット
df = pd.DataFrame(
    iris.data, 
    columns = iris.feature_names
)

# 目的変数をspeciesカラムに格納
df['species'] = iris.target

# virgicolorとvirginicaを抜き出す
df = df[df['species'].isin([1, 2])]

df
```

   * sepal_lengthとpetal_length

```python
# sepal_lengthとpetal_lengthを抜き出す
df = df.iloc[:, [0, 2, 4]]

df
```

# 【問題2】前処理・学習用データと検証用データの分割
* 前処理で学習用データ(75%)と検証用データ(25%)への分割を行う
* scikit-learnのtrain_test_splitを利用
* 特徴量（説明変数）をX、正解（目的変数）をyというndarrayに格納

```python
from sklearn.model_selection import train_test_split

# 説明変数・目的変数をX・yに格納
X = df.iloc[:, [0, 1]].values
y = df.iloc[:, [2]].values

# 分割
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.25,
    random_state = 15
)

print('X_train：')
print(X_train)
print('X_test：')
print(X_test)
print('y_train：')
print(y_train)
print('y_test：')
print(y_test)
```

# 【問題3】前処理・標準化
* 標準化を行う
* scikit-learnのStandardScalerを利用
* 標準化は学習用データでfitを行い、学習用、検証用双方のデータにtransformを行う

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# 学習用データの最大値、最小値、平均、標準偏差、傾き...などの統計を取得
scaler.fit(X_train)

# 統計に基づき各データを書き換え
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

print('X_train：')
print(X_train)
print('X_test：')
print(X_test)
```

# 【問題4】学習と推定
* 最近傍法で学習を行う
* scikit-learnのKNeighborsClassifierを利用
* 学習・検証用データをpredictメソッドに入力し推定
* n_neighborsの数にデフォルトの5、および1と3を指定

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# 予測結果格納用
y_pred_ary = np.zeros((0, 25))

# n_neighborsに指定する数
n_neighbors = [1, 3, 5]

for n in n_neighbors:
    neigh = KNeighborsClassifier(n)
   
    # 学習用データから学習
    neigh.fit(X_train, y_train.reshape(-1)) # エラーが出るためreshape
    
    #  予測を実施・格納
    y_pred_ary = np.append(y_pred_ary, [neigh.predict(X_test)], axis = 0)
    
    
print(y_pred_ary)
```

# 【問題5】評価
* scikit-learnのmetricsを使って、評価指標の計算を行う
* 指標は下記の通り

    TP: True Positive  
    正しく陽性と判定できた場合

    TP: True Negative  
    正しく陰性と判定できた場合

    FP: False Positive  
    本来は陰性なところを、誤って陽性と判定してしまった場合

    FN: False Negative  
    本来は陽性なところを、誤って陰性と判定してしまった場合　　
  
<br><br>
 
 * accuracyを求める 計算式は下記の通り　　
 
    $Accuracy = \frac{TP + TN}{TP + FP + FN + TN}$

```python
from sklearn import metrics

for y_pred in y_pred_ary:
    print('accuracy : {}'.format(metrics.accuracy_score(y_test, y_pred)))
```

 * precisionを求める 計算式は下記の通り　　
 
    $precision = \frac{TP}{TP + FP}$

```python
for y_pred in y_pred_ary:
    print('precision : {}'.format(metrics.precision_score(y_test, y_pred)))
```

 * recallを求める 計算式は下記の通り　　
 
    $recall = \frac{TP}{TP + FN}$

```python
for y_pred in y_pred_ary:
    print('recall : {}'.format(metrics.recall_score(y_test, y_pred)))
```

* f値を求める 計算式は下記の通り  
    $F-measure = 2 \frac{Precision \cdot Recall}{Precision + Recall}$

```python
for y_pred in y_pred_ary:
    print('f値 : {}'.format(metrics.f1_score(y_test, y_pred)))
```

* 混合行列  



```python
for y_pred in y_pred_ary:
    print(metrics.confusion_matrix(y_test, y_pred))
```

# 【問題6】可視化

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches

def decision_region(X_train, y_train, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は学習用データである。

    Parameters
    ----------------
    X_train : ndarray, shape(n_samples, 2)
        学習用データの特徴量
    y_train : ndarray, shape(n_samples,)
        学習用データの正解値
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X_train[:,0])-0.5, np.max(X_train[:,0])+0.5, step), np.arange(np.min(X_train[:,1])-0.5, np.max(X_train[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y_train)):
        plt.scatter(X_train[y_train==target][:, 0], X_train[y_train==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
    
    
# 3-nnでインスタンス化
neigh = KNeighborsClassifier(3)
   
# 学習用データから学習
neigh.fit(X_train, y_train.reshape(-1))   # エラーが出るためreshape
    
# プロット関数をコール
decision_region(
    X_train=X_train, 
    y_train=y_train.reshape(-1),                # エラーが出るためreshape
    model=neigh,
    title='decision region of 3-nn', 
    xlabel='sepal length', 
    ylabel='petal length'
)
```
